```python
dfs = pd.read_html(link)
dfs
```

```python
df=dfs[0]
df
```

```python
empty = 'Not assigned'
df = df[(df.Postcode != empty ) & (df.Borough != empty) & (df.Neighbourhood != empty)]
```

```python
df.reset_index()
```

```python
def Neighbourhood_list(grouped):    
    return ', '.join(sorted(grouped['Neighbourhood'].tolist()))
                    
grp = df.groupby(['Postcode', 'Borough'])
df = grp.apply(Neighbourhood_list).reset_index(name='Neighbourhood')
```
